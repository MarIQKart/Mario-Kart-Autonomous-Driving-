## Video to Image Data

In this notebook, each video file is streamed using OpenCV, and each frame of the video is saved as a separate file. Since each video was recorded and processed at separate times, instead of writing the program to perform all of the video streaming in batch, each separate video has its own copy/paste cell for processing the dataset.

### Process the Video: Driving Center of the Road

```python
# === Necessary Import === #
import cv2

# === Open Video File for Streaming === #
vidcap        = cv2.VideoCapture( 'center.mp4' )
success,image = vidcap.read( )
count         = 0

# === Iteratively Process the Video Frame-by-Frame === #
while success:
    cv2.imwrite( 'dataset/center/frame{}.jpg'.format( count ) , image )  # save frame
    success , image = vidcap.read( )  # get next frame
    print( 'Read a new frame: ' , success )  # logging
    count += 1
    
# === Report When Finished === #
print( 'Done' )
```
