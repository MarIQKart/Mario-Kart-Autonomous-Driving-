```python
import cv2
vidcap = cv2.VideoCapture('center.mp4')
success,image = vidcap.read()
count = 0
while success:
    cv2.imwrite('dataset/center/frame{}.jpg'.format(count), image)
    success,image = vidcap.read()
    print('Read a new frame: ',success)
    count += 1
print('Done')
```
